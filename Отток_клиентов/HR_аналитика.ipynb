# Оптимизация управления персоналом для HR-аналитиков

## Описание проекта

HR-аналитики компании «Работа с заботой» помогают бизнесу оптимизировать управление персоналом: бизнес предоставляет данные, а аналитики предлагают, как избежать финансовых потерь и оттока сотрудников. В этом HR-аналитикам пригодится машинное обучение, с помощью которого получится быстрее и точнее отвечать на вопросы бизнеса.
Компания предоставила данные с характеристиками сотрудников компании. Среди них — уровень удовлетворённости сотрудника работой в компании. Эту информацию получили из форм обратной связи: сотрудники заполняют тест-опросник, и по его результатам рассчитывается доля их удовлетворённости от 0 до 1, где 0 — совершенно неудовлетворён, 1 — полностью удовлетворён. 
Собирать данные такими опросниками не так легко: компания большая, и всех сотрудников надо сначала оповестить об опросе, а затем проследить, что все его прошли. 
**Первая задача** — построить модель, которая сможет предсказать уровень удовлетворённости сотрудника на основе данных заказчика. 
**Вторая задача** — построить модель, которая сможет на основе данных заказчика предсказать то, что сотрудник уволится из компании.
**Почему бизнесу это важно**: удовлетворённость работой напрямую влияет на отток сотрудников. А предсказание оттока — одна из важнейших задач HR-аналитиков. Внезапные увольнения несут в себе риски для компании, особенно если уходит важный сотрудник.

## Описание данных

**id** — уникальный идентификатор сотрудника;

**dept** — отдел, в котором работает сотрудник;

**level** — уровень занимаемой должности;

**workload** — уровень загруженности сотрудника;

**employment_years** — длительность работы в компании (в годах);

**last_year_promo** — показывает, было ли повышение за последний год;

**last_year_violations** — показывает, нарушал ли сотрудник трудовой договор за последний год;

**supervisor_evaluation** — оценка качества работы сотрудника, которую дал руководитель;

**salary** — ежемесячная зарплата сотрудника;

**job_satisfaction_rate** — уровень удовлетворённости сотрудника работой в компании, целевой признак.

**quit**- увольнение сотрудника в компании

### Обучение моделей

**Задача** по скольку перед нами стоит задача регресии, то в качестве моделей используем DecisionTreeRegressor,LinearRegressor, и оценим полученные предсказания при помощи метрики smape_score

```python
y_train=data_train['job_satisfaction_rate']
X_train=data_train.drop('job_satisfaction_rate',axis=1)
```

```python
y_test=df['job_satisfaction_rate']
X_test=df.drop('job_satisfaction_rate',axis=1)
```

```python
ohe_columns=['dept','last_year_promo','last_year_violations']
num_columns=['employment_years','salary']
ord_columns=['level','workload']
```

```python
ohe_pipe=Pipeline([
    ('simple_imputer_one',SimpleImputer(missing_values=np.nan,strategy='most_frequent')),
('one',OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False))
])
ord_pipe=Pipeline([
    ('simple_imputer_ord_before',SimpleImputer(missing_values=np.nan,strategy='most_frequent')),
('ord',OrdinalEncoder(categories=[['junior','midle','sinior'],['low','medium','high']],
                      handle_unknown='use_encoded_value', unknown_value=np.nan)),
    ('simple_imputer_ord_after',SimpleImputer(missing_values=np.nan,strategy='most_frequent'))
])
data_preprocessor=ColumnTransformer([
    ('one',ohe_pipe,ohe_columns),
    ('ord',ord_pipe,ord_columns),
    ('num',StandardScaler(),num_columns)
],
remainder='passthrough')
```

```python
pipe_final=Pipeline([
    ('prepross',data_preprocessor),
('models',DecisionTreeRegressor(random_state=random_state))])
```

```python
param_distributions=[
    { 'models': [LinearRegression()],
     'prepross__num': [StandardScaler(),MinMaxScaler()]
        },
    {
        'models': [DecisionTreeRegressor(random_state=random_state)],
        'models__max_depth': range (1,50),
        'models__max_features': range (1,50),
        'models__max_leaf_nodes': range (100,1000),
        
        'prepross__num': [StandardScaler(),MinMaxScaler()]
    }
]
```

```python
def smape_score(y_true,y_pred):
    if isinstance(y_true, pd.Series):
        y_true = y_true.values
    if isinstance(y_pred, pd.Series):
        y_pred = y_pred.values
    score=0
    for i in range(len(y_true)):
        score=score+(np.abs(y_true[i]-y_pred[i]))/(0.5*(np.abs(y_true[i])+np.abs(y_pred[i])))
    score=score*100/len(y_true)
    return score   
```

```python
smape_score_=make_scorer(smape_score,greater_is_better=False)
```

```python
model1=RandomizedSearchCV(pipe_final,param_distributions,cv=5,scoring=smape_score_,n_iter=100)
```

Для определения гиперпараметров мы взяли рандомизированный подбор, поскольку по времени он быстрее определяет и качество ничуть не хуже, чем у сеточного

**Альтернативная гипотеза** Уровень удовлетворенности работой влияет на увольнение сотрудника, т.е средний уровень удовлетворенности у уволенного и действующего достаточно отличается

Поскольку нам неважно в какую сторону отличается средний уровень коэффициента удовлетворенности, мы используем двуxстороннюю гипотезу для независимых выборок

```python
result=st.ttest_ind(data1['job_satisfaction_rate'],data2['job_satisfaction_rate'],alternative='two-sided')
zone=0.05
print(result.pvalue)
if result.pvalue<zone:
    print('Уровень удовлетворенности работой влияет на увольнение сотрудника,'
          'т.е средний уровень удовлетворенности у уволенного и действующего достаточно отличается')
else:
    print('Уровень удовлетворенности работой не влияет на увольнение сотрудника, т.е средний уровень удовлетворенности у уволенного и действующего одинаков')
```

Таким образом мы отвергаем нашу нулевую гипотезу, а значит коэфициент удовлетворенности влияет на целевой признак

### Обучение моделей

**Задача** Обучим несколько моделей для задачи бинарной классификации (DecisionTreeClassifier,KNeighborsClassifier,LogisticRegression) и определим лучшую

```python
ohe_columns=['dept','last_year_promo','last_year_violations']
num_columns=['employment_years','salary']
ord_columns=['level','workload']
ohe_pipe=Pipeline([
    ('simple_imputer_one',SimpleImputer(missing_values=np.nan,strategy='most_frequent')),
('one',OneHotEncoder(drop='first',handle_unknown='ignore',sparse_output=False))
])
ord_pipe=Pipeline([
    ('simple_imputer_ord_before',SimpleImputer(missing_values=np.nan,strategy='most_frequent')),
('ord',OrdinalEncoder(categories=[['junior','midle','sinior'],['low','medium','high']],
                      handle_unknown='use_encoded_value', unknown_value=np.nan)),
    ('simple_imputer_ord_after',SimpleImputer(missing_values=np.nan,strategy='most_frequent'))
])
data_preprocessor=ColumnTransformer([
    ('one',ohe_pipe,ohe_columns),
    ('ord',ord_pipe,ord_columns),
    ('num',StandardScaler(),num_columns)
],
remainder='passthrough')
pipe_final=Pipeline([
    ('prepross',data_preprocessor),
('models',LogisticRegression(random_state=random_state))])
param_distributions=[
    { 'models': [LogisticRegression(solver='saga',random_state=random_state,penalty='l1',max_iter=500)],
     'models__C': range (1,5),
     'prepross__num': [StandardScaler(),MinMaxScaler()]
        },
    {
        'models': [DecisionTreeClassifier(random_state=random_state)],
        'models__max_depth': range (10,1000),
        'models__max_features': range (1,25),
        'models__max_leaf_nodes': range (2,100),
        'prepross__num': [StandardScaler(),MinMaxScaler()]
    },
    {
        'models': [KNeighborsClassifier()],
        'models__n_neighbors': range (5,200),
        'prepross__num': [StandardScaler(),MinMaxScaler()]
    }  
]
```

```python
label=LabelEncoder()
```

```python
print(f'0 кодируется: {label.classes_[0]}')
print(f'1 кодируется: {label.classes_[1]}')
```

Важно понимать, как кодируется целевой признак

```python
model2=RandomizedSearchCV(pipe_final,param_distributions,cv=5,scoring='roc_auc',n_iter=100,random_state=random_state)
```

Будем для подбора оптимальных гиперпараметров использовать Рандомизированный поиск, поскольку по времени он определяет быстрее. В качестве метрики возьмем roc_auc, поскольку она не зависит от порога классификации и устойчива к доминированию одного класса над другим
