# Определение мошенических транзакций

## Описание проекта

Компании, выпускающие кредитные карты, могли распознавать мошеннические транзакции по кредитным картам, чтобы с клиентов не взималась плата за товары, которые они не покупали.Набор данных содержит транзакции, совершенные по кредитным картам в сентябре 2013 года европейскими держателями карт.
В этом наборе данных представлены транзакции, произошедшие за два дня, где у нас 492 мошенничества из 284 807 транзакций.
Нужно обучить модель, которая будет выявлять мошенничиские транзакции.

## Описание данных

**Time** - кол-во секунд, прошедших между этой транзакцией и первой транзакцией в наборе данных  
**V1-V28** - конфеденциальные признаки, которые не могут быть рассекречены  
**Amount** - Сумма транзакции  
**Class** - является ли операция мошеннической или нет (1 да, 0 нет)

````python
def pie_plot(data,column):
    res=data[column].value_counts()
    plt.pie(res.values,labels=res.index,autopct='%1.2f%%')
    plt.title(f'Круговая диаграмма {column}')
    plt.show()
````

````python
def boxplot_(data, columns, target):
    for column in columns:
        res=[data[data[target]==i][column] for i in data[target].unique()]
        labels=data[target].unique()
        plt.figure(figsize=(3,3))
        plt.title(f'Диаграмма размаха {column} от {target}')
        plt.boxplot(res,labels=labels)
        plt.xlabel(f'{target}')
        plt.ylabel(f'{column}')
        plt.show()
````

## Нейронная сеть

**Задача** Построить архитектуру нейронной сети и проверить ее эффективность

```
X_train=data_prep.fit_transform(X_train)
X_test=data_prep.transform(X_test)
X_train=torch.FloatTensor(X_train)
y_train=torch.FloatTensor(y_train.values).reshape(-1,1)
X_test=torch.FloatTensor(X_test)
y_test=torch.FloatTensor(y_test.values).reshape(-1,1)
class Fraud_Detection(torch.nn.Module):
  def __init__(self):
    super(Fraud_Detection,self).__init__()
    self.fc1=torch.nn.Linear(X_train.shape[1],32)
    self.act1=torch.nn.ReLU()
    self.fc2=torch.nn.Linear(32,16)
    self.act2=torch.nn.ReLU()
    self.fc3=torch.nn.Linear(16,1)
    self.act3=torch.nn.Sigmoid()
  def forward(self,x):
    x=self.fc1(x)
    x=self.act1(x)
    x=self.fc2(x)
    x=self.act2(x)
    x=self.fc3(x)
    x=self.act3(x)
    return x

fraud_net=Fraud_Detection()
class_weights = torch.FloatTensor([len(y_train[y_train == 0]) / len(y_train[y_train == 1])])
loss=torch.nn.BCELoss(weight=class_weights)
optimizer=torch.optim.Adam(fraud_net.parameters(),lr=0.0001)
batch_size=500
precision_matrix=[]
recall_matrix=[]
f1_matrix=[]
rocauc_matrix=[]
epoch_matrix=[]
for epoch in range (400):
  running_loss = 0.0
  order=np.random.permutation(len(X_train))
  for start_index in range (0,len(X_train),batch_size):
    optimizer.zero_grad()
    fraud_net.train()
    batch_index=order[start_index:start_index+batch_size]
    X_batch=X_train[batch_index]
    y_batch=y_train[batch_index]
    preds=fraud_net.forward(X_batch)
    loss_value=loss(preds,y_batch)
    loss_value.backward()
    optimizer.step()
    running_loss += loss_value.item()
  epoch_loss = running_loss / len(X_train)
  if (epoch + 1) % 10 == 0:
        fraud_net.eval()
        with torch.no_grad():
            test_preds = fraud_net(X_test)
            test_preds = (test_preds >= 0.5).float().numpy()

        precision = precision_score(y_test, test_preds)
        precision_matrix.append(precision)
        recall = recall_score(y_test, test_preds)
        recall_matrix.append(recall)
        f1 = f1_score(y_test, test_preds)
        f1_matrix.append(f1)
        roc_auc = roc_auc_score(y_test, test_preds)
        rocauc_matrix.append(roc_auc)
        epoch_matrix.append(epoch)
        print(f'Epoch: {epoch}')
        print(f'Precision: {precision:.2f}')
        print(f'loss {epoch_loss}')
        print(f'Recall: {recall:.2f}')
        print(f'F1-Score: {f1:.2f}')
        print(f'ROC-AUC: {roc_auc:.4f}')
        print('-----------')
```
